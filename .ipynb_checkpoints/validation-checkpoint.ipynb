```python
import pandas as pd
import zipfile

ZIP_FILE_PATH = 'data/export_S Ochs.zip'

with zipfile.ZipFile(ZIP_FILE_PATH, 'r') as zf:
    cgm_df    = pd.read_csv(zf.open('cgm_data_1.csv'), skiprows=1)
    bolus_df  = pd.read_csv(zf.open('Insulin data/bolus_data_1.csv'), skiprows=1)
    alarms_df = pd.read_csv(zf.open('alarms_data_1.csv'), skiprows=1)
    insulin_df = pd.read_csv(zf.open('Insulin data/insulin_data_1.csv'), skiprows=1)

cgm_df = cgm_df.rename(columns={
    'Timestamp': 'timestamp',
    'CGM Glucose Value (mg/dl)': 'glucose'
})
bolus_df = bolus_df.rename(columns={
    'Timestamp': 'timestamp',
    'Insulin Delivered (U)': 'bolus',
    'Carbs Input (g)': 'carbs',
})
alarms_df = alarms_df.rename(columns={'Timestamp': 'timestamp', 'Alarm/Event': 'event'})
insulin_df = insulin_df.rename(columns={
    'Timestamp': 'timestamp',
    'Total Bolus (U)': 'total_bolus',
    'Total Basal (U)': 'total_basal',
    'Total Insulin (U)': 'total_insulin',
})

cgm_df['timestamp']    = pd.to_datetime(cgm_df['timestamp'],    utc=True)
bolus_df['timestamp']  = pd.to_datetime(bolus_df['timestamp'],  utc=True)
alarms_df['timestamp'] = pd.to_datetime(alarms_df['timestamp'], utc=True)
insulin_df['timestamp'] = pd.to_datetime(insulin_df['timestamp'], utc=True)

cgm_df   = cgm_df.sort_values('timestamp').reset_index(drop=True)
bolus_df = bolus_df.sort_values('timestamp').reset_index(drop=True)

print(f'Imported {len(cgm_df)} CGM readings and {len(bolus_df)} bolus events.')
```

```python
pod_failures = alarms_df[alarms_df['event'].str.contains('delivery stopped', case=False, na=False)].copy()
pod_failure_times = pod_failures['timestamp']

print(f"Found {len(pod_failure_times)} pod delivery failure event(s):")
for _, row in pod_failures.iterrows():
    print(f"  {row['timestamp'].strftime('%Y-%m-%d %H:%M')}  {row['event']}")
```

```python
events_df = pd.merge_asof(
    bolus_df,
    cgm_df[['timestamp', 'glucose']],
    on='timestamp',
    direction='backward',
    tolerance=pd.Timedelta('10 minutes')
)

events_df = events_df[events_df['bolus'] > 0].copy()

print(f"Successfully merged {len(events_df)} bolus events.")
print(events_df.head())
```

```python
TZ_OFFSET_HOURS = 0  # Glooko exports in local time — no UTC conversion needed

meal_boluses = events_df[events_df['carbs'].fillna(0) > 0].copy()
meal_boluses['local_hour'] = (
    meal_boluses['timestamp'].dt.hour +
    meal_boluses['timestamp'].dt.minute / 60 +
    TZ_OFFSET_HOURS
) % 24

def time_slot(h):
    if 5  <= h < 11: return 'Breakfast (5–10) '
    if 11 <= h < 16: return 'Lunch    (11–15)'
    if 16 <= h < 22: return 'Dinner   (16–21)'
    return                  'Night    (22–4) '

meal_boluses['slot'] = meal_boluses['local_hour'].apply(time_slot)

habits = meal_boluses.groupby('slot').agg(
    events   =('carbs',      'count'),
    avg_carbs=('carbs',      'mean'),
    avg_bolus=('bolus',      'mean'),
    med_hour =('local_hour', 'median'),
).sort_values('med_hour')

print("── Inferred Meal Habits ──\n")
print(f"  {'Slot':<20} {'Events':>6}  {'Avg Carbs':>9}  {'Avg Bolus':>9}  {'Typical Time':>12}")
print("  " + "-" * 63)
for slot, row in habits.iterrows():
    h   = int(row['med_hour'])
    m   = int((row['med_hour'] % 1) * 60)
    ampm = 'am' if h < 12 else 'pm'
    h12  = h if h <= 12 else h - 12
    print(f"  {slot:<20} {int(row['events']):>6}  {row['avg_carbs']:>8.0f}g  {row['avg_bolus']:>8.1f}U  {h12:>8}:{m:02d}{ampm}")
```

```python
import numpy as np
from datetime import timedelta

EVALUATION_WINDOW_HOURS = 4

def evaluate_event_window(row, all_events_df, cgm_df, pod_failure_times):
    start_time = row['timestamp']
    end_time = start_time + timedelta(hours=EVALUATION_WINDOW_HOURS)

    # Pod delivery interrupted during window — basal was cut off, outcome is confounded
    failures_in_window = pod_failure_times[
        (pod_failure_times >= start_time) & (pod_failure_times <= end_time)
    ]
    if not failures_in_window.empty:
        return pd.Series({'status': 'Rejected: Pod Delivery Interrupted', 'end_glucose': np.nan, 'min_glucose': np.nan})

    subsequent_boluses = all_events_df[
        (all_events_df['timestamp'] > start_time) &
        (all_events_df['timestamp'] <= end_time) &
        (all_events_df['bolus'] > 0)
    ]

    if not subsequent_boluses.empty:
        # New meal within window — unrelated to first bolus performance
        if subsequent_boluses['carbs'].fillna(0).sum() > 0:
            return pd.Series({'status': 'Rejected: New Meal in Window', 'end_glucose': np.nan, 'min_glucose': np.nan})
        # Correction-only follow-up — first bolus undershot
        else:
            return pd.Series({'status': 'Clean Window', 'end_glucose': 999.0, 'min_glucose': np.nan})

    window_cgm = cgm_df[
        (cgm_df['timestamp'] >= start_time) &
        (cgm_df['timestamp'] <= end_time)
    ]

    if window_cgm.empty:
        return pd.Series({'status': 'Rejected: Missing CGM Data', 'end_glucose': np.nan, 'min_glucose': np.nan})

    end_glucose = window_cgm.iloc[-1]['glucose']
    min_glucose = window_cgm['glucose'].min()

    status = 'Clean Window'
    if min_glucose < 70:
        status = 'Rejected: Hypoglycemic Event Detected'

    return pd.Series({'status': status, 'end_glucose': end_glucose, 'min_glucose': min_glucose})

outcomes = events_df.apply(evaluate_event_window, axis=1, args=(events_df, cgm_df, pod_failure_times))
processed_events = pd.concat([events_df, outcomes], axis=1)

print("Event status breakdown:")
print(processed_events['status'].value_counts())
```

```python
# Filter for only the mathematically clean evaluation windows
clean_events = processed_events[processed_events['status'] == 'Clean Window'].copy()

# Categorize the outcome of the clean events based on target ranges
def grade_ratio(row):
    target_high = 130
    target_low = 80

    if row['end_glucose'] > target_high:
        return 'Ratio Too Weak (Failed High)'
    elif row['end_glucose'] < target_low:
        return 'Ratio Too Strong (Failed Low)'
    else:
        return 'Ratio Accurate (In Range)'

clean_events['evaluation'] = clean_events.apply(grade_ratio, axis=1)

# Separate into Meals (Testing ICR) and Corrections (Testing ISF)
meals = clean_events[clean_events['carbs'] > 0]
corrections = clean_events[(clean_events['carbs'] == 0) & (clean_events['bolus'] > 0)]

print("--- Meal Ratio (ICR) Performance ---")
print(meals['evaluation'].value_counts())

print("\n--- Correction Factor (ISF) Performance ---")
print(corrections['evaluation'].value_counts())
```

```python
# ── Configuration ────────────────────────────────────────────────────────────
TARGET_GLUCOSE     = 110   # mg/dL
CONFIGURED_ICR     = 10    # g per unit
CONFIGURED_ISF     = 30    # mg/dL per unit
REVERSE_CORRECTION = True
TZ_OFFSET_HOURS    = 0     # Glooko exports in local time — no UTC conversion needed

# ── Shortfall helper ──────────────────────────────────────────────────────────
# For any event that failed high, compute additional units that would have been
# needed to reach TARGET using CONFIGURED_ISF. Derive implied ratio from
# (what was given) + (what was missing).

def implied_isf_from_shortfall(row):
    bg = row['Blood Glucose Input (mg/dl)']
    if bg <= 0 or row['end_glucose'] >= 900:
        return np.nan
    required_drop   = bg - TARGET_GLUCOSE
    actual_drop     = bg - row['end_glucose']
    shortfall_units = (required_drop - actual_drop) / CONFIGURED_ISF
    total_needed    = row['bolus'] + shortfall_units
    return required_drop / total_needed if total_needed > 0 else np.nan

def implied_icr_from_shortfall(row):
    bg = row['Blood Glucose Input (mg/dl)'] if row['Blood Glucose Input (mg/dl)'] > 0 else TARGET_GLUCOSE
    correction_component = (bg - TARGET_GLUCOSE) / CONFIGURED_ISF
    carb_bolus_given = row['bolus'] - correction_component
    if row['end_glucose'] >= 900 or np.isnan(row['end_glucose']):
        return np.nan
    shortfall_units       = max(0, row['end_glucose'] - TARGET_GLUCOSE) / CONFIGURED_ISF
    total_carb_bolus_needed = carb_bolus_given + shortfall_units
    return row['carbs'] / total_carb_bolus_needed if total_carb_bolus_needed > 0 else np.nan

# ── ISF from corrections ──────────────────────────────────────────────────────
corr = clean_events[clean_events['carbs'].fillna(0) == 0].copy()
corr_n_total   = len(corr)
corr_n_high    = (corr['end_glucose'] > TARGET_GLUCOSE).sum()
corr_n_oor     = corr['evaluation'].eq('Ratio Too Weak (Failed High)').sum()
corr_n_ok      = corr['evaluation'].eq('Ratio Accurate (In Range)').sum()
corr_n_stack   = (corr['end_glucose'] >= 900).sum()

corr['implied_isf'] = corr.apply(implied_isf_from_shortfall, axis=1)
inferred_isf = corr['implied_isf'].median()

# ── ICR from meals ────────────────────────────────────────────────────────────
meal = clean_events[clean_events['carbs'].fillna(0) > 0].copy()
meal_n_total   = len(meal)
meal_n_high    = (meal['end_glucose'] > TARGET_GLUCOSE).sum()
meal_n_oor     = meal['evaluation'].eq('Ratio Too Weak (Failed High)').sum()
meal_n_ok      = meal['evaluation'].eq('Ratio Accurate (In Range)').sum()

meal['implied_icr'] = meal.apply(implied_icr_from_shortfall, axis=1)
inferred_icr = meal['implied_icr'].median()

# ── Time-of-day breakdown ─────────────────────────────────────────────────────
clean_events['local_hour'] = (clean_events['timestamp'].dt.hour + TZ_OFFSET_HOURS) % 24

def time_slot(hour):
    if 5  <= hour < 11: return 'Breakfast (5–10) '
    if 11 <= hour < 16: return 'Lunch    (11–15)'
    if 16 <= hour < 22: return 'Dinner   (16–21)'
    return                     'Night    (22–4) '

clean_events['slot'] = clean_events['local_hour'].apply(time_slot)
tod = clean_events.groupby('slot')['evaluation'].value_counts().unstack(fill_value=0)

# ── Print recommendations ─────────────────────────────────────────────────────
print("=" * 52)
print("  LAGOM RECOMMENDATIONS")
print("=" * 52)

print(f"\n── Insulin Sensitivity (ISF) ──")
print(f"  Configured  : {CONFIGURED_ISF} mg/dL per unit")
print(f"  Outcome     : {corr_n_high}/{corr_n_total} trended high (>{TARGET_GLUCOSE})  |  {corr_n_oor} out of range (>130)  ({corr_n_stack} required stacking)")
if not np.isnan(inferred_isf):
    print(f"  Implied ISF : ~{inferred_isf:.1f} mg/dL per unit")
    if inferred_isf < CONFIGURED_ISF - 2:
        suggested = round(inferred_isf / 5) * 5
        print(f"  → Consider lowering ISF to ~{suggested} mg/dL per unit.")
    elif inferred_isf > CONFIGURED_ISF + 2:
        suggested = round(inferred_isf / 5) * 5
        print(f"  → Consider raising ISF to ~{suggested} mg/dL per unit.")
    else:
        print(f"  → ISF appears well-calibrated.")

print(f"\n── Insulin-to-Carb Ratio (ICR) ──")
print(f"  Configured  : {CONFIGURED_ICR} g per unit")
print(f"  Outcome     : {meal_n_high}/{meal_n_total} trended high (>{TARGET_GLUCOSE})  |  {meal_n_oor} out of range (>130)")
if not np.isnan(inferred_icr):
    print(f"  Implied ICR : ~{inferred_icr:.1f} g per unit")
    if inferred_icr < CONFIGURED_ICR - 0.4:
        suggested = round(inferred_icr / 0.5) * 0.5
        print(f"  → Consider lowering ICR to ~{suggested:.1f} g per unit.")
    elif inferred_icr > CONFIGURED_ICR + 0.4:
        suggested = round(inferred_icr / 0.5) * 0.5
        print(f"  → Consider raising ICR to ~{suggested:.1f} g per unit.")
    else:
        print(f"  → ICR appears well-calibrated.")

print(f"\n── Time-of-Day Breakdown ──")
for slot in sorted(tod.index):
    row = tod.loc[slot]
    high = row.get('Ratio Too Weak (Failed High)', 0)
    ok   = row.get('Ratio Accurate (In Range)', 0)
    low  = row.get('Ratio Too Strong (Failed Low)', 0)
    print(f"  {slot}  high={high}  ok={ok}  low={low}")
print()
```

```python
insulin_df['date'] = insulin_df['timestamp'].dt.date
insulin_df['bolus_pct'] = insulin_df['total_bolus'] / insulin_df['total_insulin'] * 100
insulin_df['basal_pct'] = insulin_df['total_basal'] / insulin_df['total_insulin'] * 100

failure_dates = set(pod_failures['timestamp'].dt.date)
insulin_df['note'] = insulin_df['date'].apply(lambda d: '<- pod failure' if d in failure_dates else '')

print("── Daily Insulin Totals ──\n")
print(f"  {'Date':<12} {'Basal':>8} {'Bolus':>8} {'TDD':>8}  {'Basal%':>7}  {'Bolus%':>7}  Note")
print("  " + "-" * 72)
for _, row in insulin_df.sort_values('date').iterrows():
    note = f"  {row['note']}" if row['note'] else ''
    print(f"  {str(row['date']):<12} {row['total_basal']:>7.1f}U {row['total_bolus']:>7.1f}U {row['total_insulin']:>7.1f}U  {row['basal_pct']:>6.0f}%  {row['bolus_pct']:>6.0f}%{note}")

clean_days = insulin_df[insulin_df['note'] == '']
print(f"\n  Average (clean days): TDD {clean_days['total_insulin'].mean():.1f}U  |  Basal {clean_days['basal_pct'].mean():.0f}%  /  Bolus {clean_days['bolus_pct'].mean():.0f}%")
```
